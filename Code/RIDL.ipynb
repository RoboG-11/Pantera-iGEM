# Source code

```python
def dw_dt(a, w, c, mu, xi, t):
    return ((a * w) / (w + c) - (mu + xi * (( w + c )))) * w

# Parameters
a         = 20
mu        = 0.3
xi        = 0.2
t_bar     = 2.5
g_star    = (a - mu) ** 2 / (4 * xi * a)
c_star    = (-a + (np.sqrt(a ** 2 + 4 * a * (a - mu)))) / (2 * xi)
c         = 32
t_star    = ((a + (xi * c)) / (a - mu)) * t_bar

# Time Parameters
t_int  = 0
h      = 0.0001
t_max  = 10
ite    = int((t_max - t_int) / h )

# Initial conditions
w    = np.zeros(ite + 1)
w[0] = 10
t    = np.linspace(t_int, t_max, ite + 1)

# Simulation
for i in range(ite):
    w[i + 1] = w[i] + h * dw_dt(a, w[i], c, mu, xi, t)

plt.plot(t,w)
plt.xlabel("Time")
plt.ylabel("Wild mosquitoes")
plt.title("Test 1")
```
